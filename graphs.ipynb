### Importem les llibreries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import ptitprince as pt
```

```python
# Llegim el fitxer i el guardem en la variable data
data = pd.read_csv("penguins_size.csv")
```

### Pie chart

```python
# Calculem la freqüència de les diferents espècies
sps_counts = data['species'].value_counts()

# Les ordenem de menys a més, de tal manera que el gràfic mostra les freqüències de més a menys en sentit horari.
sorted_sps = sps_counts.sort_values(ascending=True).index

colors = ['#f3bc00', '#00af82', '#0075a4']

# Representem el gràfic
plt.pie(sps_counts[sorted_sps], labels=sorted_sps, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title("Distribució de les espècies de pingüins en l’Arxipèlag de Palmer (Antàrtida)")
plt.show()
```

#### Pie chart (bad practices)

```python
# Llegim el fitxer i el guardem en la variable net
net = pd.read_csv("netflix_titles.csv")
```

```python
net_counts = net['rating'].value_counts()

sorted_net = net_counts.sort_values(ascending=True).index
plt.figure(figsize=(8, 8))

plt.pie(net_counts[sorted_net], labels=sorted_net, autopct='%1.1f%%', startangle=90, labeldistance=1.1)

plt.title("Distribució de les valoracions dels shows i pel·lícules")
plt.show()
```

### Raincloud Plot

```python
palette = ['#0075a4', '#f3bc00', '#00af82']

# Creem el raincloud plot
plt.figure(figsize=(10, 6))
ax = pt.RainCloud(x='species', y='body_mass_g', data=data, palette=palette)

# Títol i etiquetes
plt.xlabel("")
plt.ylabel("Massa corporal (g)")
plt.title("Massa corporal de les espècies de pingüins en l’Arxipèlag de Palmer (Antàrtida)")

plt.show()
```
